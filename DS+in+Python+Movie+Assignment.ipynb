Find out the number of Null values in all the columns and rows. Also, find the percentage of Null values in each column. Round off the percentages upto two decimal places.

```python
# Write your code for row-wise null count here
movie.isnull().sum(axis=1).sort_values(ascending=False)
```

```python
# Write your code for column-wise null percentages here
   
def missing_coldata(d1):
    missing_data1 = pd.DataFrame(round(d1.isnull().sum().sort_values(ascending=False)/d1.shape[0]*100,2), columns=['% of missing value'])
    missing_data1['Count of Missing Values'] = d1.isnull().sum()
    return missing_data1
missing_coldata(movie)
```

```python
# Write your code for row-wise null percentages here
   
def missing_rowdata(d1):
    missing_data = pd.DataFrame(round(d1.isnull().sum(axis=1).sort_values(ascending=False)/d1.shape[1]*100,2), columns=['% of missing value'])
    missing_data['Count of Missing Values'] = d1.isnull().sum(axis=1)
    return missing_data
missing_rowdata(movie)
```
